```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cols=['fLength','fWidth','fSize','fConc','fConc1','fAsym','fM3Long','fM3Trans','fAlpha','fDist','class']
df=pd.read_csv('magic04.data',names=cols)
df.head()
```

```python
df['class'].unique()
```

```python
df['class']=(df['class']=='g').astype(int)
df.head(10)
```

```python
df
```

```python
for i in cols[:-1]:
    plt.hist(df[df['class']==1][i],color='purple',label='gama',alpha=0.7,density=True)   
    plt.hist(df[df['class']==0][i],color='yellow',label='hadron',alpha=0.7,density=True) 
    plt.title(i)  
    plt.ylabel('probability')
    plt.xlabel(i)
    plt.legend()
    plt.show()
```
